# Cleaning Data

```python
# dropping the duplicates
df.drop_duplicates()
```

```python
df.duplicated()
```

```python
# Changing date to date type and grouping my data

df['Event.Date'] = pd.to_datetime(df['Event.Date'], errors='coerce')
df['Event.Date'].isna().sum
```

```python
# Grouping Categorical columns and numerical columns
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
```

# Handling Missing Data

```python
# Checking for missing data

df.isna().sum()
```

```python
# replace placeholders with actual NaN
df.replace(['N/A', 'NA', 'Na', 'NONE', 'Unknown', '--', ''], pd.NA, inplace=True)
```

```python
# dropping rows missing critical identifiers fields
df.dropna(subset=['Make','Model'], inplace=True)
```

```python
# Replacing missing categorical data with 'unknown'
for col in categorical_cols:
    df[col].fillna('Unknown', inplace=True)
```

```python
# fill missing by grouping the make and model
df['Number.of.Engines'] = df.groupby(['Make','Model'])['Number.of.Engines'].transform(lambda x: x.fillna(x.median()))
df['Number.of.Engines'].fillna(df['Number.of.Engines'].median(), inplace=True)
df['Number.of.Engines'].isna().sum()
```

```python
injuries = ['Total.Fatal.Injuries', 'Total.Serious.Injuries', 
           'Total.Minor.Injuries', 'Total.Uninjured']
for col in injuries:
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)

```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.columns
```

```python
#Standardizing Categorical Data
exclude = ['Event.Id', 'Accident.Number','Registration.Number', 'Model' ]
for col in [c for c in categorical_cols if c not in exclude]:
    df[col] = (
        df[col]
        .astype(str)                
        .str.strip()                
        .str.title()                
    )

```

# Aircraft Safety and Risk Ranking

```python
#giving each model the total number of incidents and summed injuries
grouped = df.groupby(['Make', 'Model']).agg({
    'Total.Fatal.Injuries': 'sum',
    'Total.Serious.Injuries': 'sum',
    'Total.Minor.Injuries': 'sum',
    'Total.Uninjured': 'sum',
    'Aircraft.damage': 'count'   
}).rename(columns={'Aircraft.damage': 'Total_Accidents'}).reset_index()

grouped
```

```python
grouped['Fatality_Rate'] = grouped['Total.Fatal.Injuries'] / grouped['Total_Accidents']
grouped['Severity_Index'] = (
    (grouped['Total.Fatal.Injuries'] * 3) +
    (grouped['Total.Serious.Injuries'] * 2) +
    (grouped['Total.Minor.Injuries'] * 1)
) / grouped['Total_Accidents']
```

```python
grouped['Fatality_Rate']
```

```python
grouped['Severity_Index']
```

```python
grouped['Safety_Rank'] = grouped['Severity_Index'].rank(method='dense', ascending=True)
grouped['Safety_Rank']
```

```python
grouped
```

```python
import matplotlib.pyplot as plt
top10_safest = grouped.head(20)
top10_safest['Aircraft'] = top10_safest['Make'] + ' ' + top10_safest['Model'].astype(str)

# --- Plot ---
plt.figure(figsize=(12, 6))
plt.barh(top10_safest['Aircraft'], top10_safest['Severity_Index'], color='#81D8D0')
plt.gca().invert_yaxis()  # safest on top
plt.title('Top 10 Safest Aircraft Models (Lowest Severity Index)', fontsize=14, fontweight='bold')
plt.xlabel('Severity Index (lower = safer)', fontsize=12)
plt.ylabel('Aircraft Model', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
