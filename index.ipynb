# Handling Missing Data

```python
# Checking for missing data

df.isna().sum()
```

```python
# replace placeholders with actual NaN
df.replace(['N/A', 'NA', 'Na', 'NONE', 'Unknown', '--', ''], pd.NA, inplace=True)
```

```python
# dropping rows missing critical identifiers fields
df.dropna(subset=['Make','Model'], inplace=True)
```

```python
# Replacing missing categorical data with 'unknown'
for col in categorical_cols:
    df[col].fillna('Unknown', inplace=True)
```

```python
# fill missing no.of.engines by grouping the make and model this preserves manufacturers level patterns
df['Number.of.Engines'] = df.groupby(['Make','Model'])['Number.of.Engines'].transform(lambda x: x.fillna(x.median()))
df['Number.of.Engines'].fillna(df['Number.of.Engines'].median(), inplace=True)
df['Number.of.Engines'].isna().sum()
```

```python
injuries = ['Total.Fatal.Injuries', 'Total.Serious.Injuries', 
           'Total.Minor.Injuries', 'Total.Uninjured']
for col in injuries:
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)

```

```python
#Standardizing Categorical Data
exclude = ['Event.Id', 'Accident.Number','Registration.Number', 'Model' ]
for col in [c for c in categorical_cols if c not in exclude]:
    df[col] = (
        df[col]
        .astype(str)                
        .str.strip()                
        .str.title()                
    )

```

# Data Analysis

Objectives:
Your company wants to start operating aircraft for commercial and private use but lacks aviation experience.
We’ll analyze accident data to identify:

1. Which aircraft are the safest (lowest risk)?

2. What factors increase or decrease accident severity?

3. What operational strategies reduce exposure to high-risk conditions?

# Which Aircraft Are Lowest Risk?

```python
# Aircraft Risk Analysis: Make + Model

import pandas as pd
import numpy as np

# Clean and prepare base columns
df['Make'] = df['Make'].astype(str).str.strip().str.title()
df['Model'] = df['Model'].astype(str).str.strip().str.upper()
df['Aircraft.damage'] = df['Aircraft.damage'].astype(str).str.strip().str.title()


# Create helper analysis columns


# (a) Severity Index (Weighted injury score)
df['Severity_Index'] = (
    df['Total.Fatal.Injuries'].fillna(0)*3 +
    df['Total.Serious.Injuries'].fillna(0)*2 +
    df['Total.Minor.Injuries'].fillna(0)*1
)

# (b) Any Fatal? 
df['Any_Fatal'] = np.where(df['Total.Fatal.Injuries'] > 0, 1, 0)

# (c) Map Aircraft Damage → numeric scale
damage_map = {
    'Destroyed': 3,
    'Substantial': 2,
    'Minor': 1,
    'None': 0
}
df['Damage_Score'] = df['Aircraft.damage'].map(damage_map).fillna(1)  

# 3. Aggregate risk metrics by Make + Model

model_risk = (
    df.groupby(['Make', 'Model'], dropna=False)
      .agg(
          Total_Accidents=('Event.Id', 'count'),
          Mean_Severity=('Severity_Index', 'mean'),
          Fatal_Rate=('Any_Fatal', 'mean'),
          Mean_Damage=('Damage_Score', 'mean')
      )
      .reset_index()
)

# Filter for meaningful sample size (>=15 accidents)
model_risk = model_risk[model_risk['Total_Accidents'] >= 15].copy()


# 4. Normalize metrics and compute Composite Risk

def minmax(series):
    smin, smax = series.min(), series.max()
    if smax == smin:
        return pd.Series(0, index=series.index)
    return (series - smin) / (smax - smin)

for c in ['Mean_Severity', 'Fatal_Rate', 'Mean_Damage']:
    model_risk[f'Norm_{c}'] = minmax(model_risk[c])

# Average normalized values → Composite Risk Score
model_risk['Composite_Risk'] = model_risk[
    ['Norm_Mean_Severity', 'Norm_Fatal_Rate', 'Norm_Mean_Damage']
].mean(axis=1)

# 5. Final presentation fields for Tableau
model_risk['Fatal_Rate_Pct']   = (model_risk['Fatal_Rate'] * 100).round(2)
model_risk['Composite_Risk_r'] = model_risk['Composite_Risk'].round(3)
model_risk['Make_Model']       = model_risk['Make'] + " " + model_risk['Model']

# Top 10 safest aircraft
lowest_risk = (
    model_risk.sort_values('Composite_Risk', ascending=True)
              .head(10)
              .loc[:, ['Make', 'Model', 'Total_Accidents',
                       'Mean_Severity', 'Fatal_Rate_Pct',
                       'Mean_Damage', 'Composite_Risk_r']]
)

print("\nTop 10 Safest Aircraft Models:")
print(lowest_risk.to_string(index=False))

# 6. Export clean CSVs for Tableau

model_risk.to_csv('aircraft_summary.csv', index=False)
lowest_risk.to_csv('aircraft_summary_top10.csv', index=False)

```

# Findings
Grouping accident data by Make + Model reveals that composite risk (combined average of severity, fatality rate, and damage) is lowest for newer, large-transport jets such as the Boeing 747, 787, 737-7H4 and the Bombardier CL-600-2B19 regional jet. while smaller or older models (e.g., McDonnell Douglas DC-9, MD-80) show higher severity averages.

# Recommendation 1
Begin operations with modern transport-category aircraft (e.g., Boeing 737-800 or 787) for initial fleet acquisition.
These models show the lowest historical accident severity, suggesting the best risk-to-profit ratio for a new operator.

# Weather Conditions and Accident Severity

```python
# Weather Analysis
weather_summary = (
    df.groupby('Weather.Condition')
      .agg(
          Total_Accidents=('Event.Id', 'count'),
          Mean_Severity=('Severity_Index', 'mean'),
          Fatal_Rate=('Any_Fatal', 'mean'),
          Mean_Damage=('Damage_Score', 'mean')
      )
      .reset_index()
      .sort_values('Mean_Severity', ascending=False)
)

weather_summary
```

```python
weather_summary.to_csv("factor_Weather_Condition.csv", index=False)
```

# Findings
Accidents occurring in IMC Weather conditions are roughly four times more severe and three-to-four times more likely to be fatal than those in VMC Weather conditions. Mean Damage also increases, indicating that aircraft in IMC crashes suffer more destruction.

# Recommendation 2
Invest early in IMC-capable aircraft, IFR-rated pilots, and weather-driven dispatch policies.

# How the phase of flight and purpose of flight relate to accident severity and fatality rates.

```python
# Accident severity by phase of flight
phase_summary = (
    df.groupby('Broad.phase.of.flight')
      .agg(
          Total_Accidents=('Event.Id', 'count'),
          Mean_Severity=('Severity_Index', 'mean'),
          Fatal_Rate=('Any_Fatal', 'mean'),
          Mean_Damage=('Damage_Score', 'mean')
      )
      .reset_index()
      .sort_values('Mean_Severity', ascending=False)
)

phase_summary
```

```python
# Accident Severity by Purpose of flight
purpose_summary = (
    df.groupby('Purpose.of.flight')
      .agg(
          Total_Accidents=('Event.Id', 'count'),
          Mean_Severity=('Severity_Index', 'mean'),
          Fatal_Rate=('Any_Fatal', 'mean'),
          Mean_Damage=('Damage_Score', 'mean')
      )
      .reset_index()
      .sort_values('Mean_Severity', ascending=False)
)

purpose_summary['Purpose.of.flight'] = purpose_summary['Purpose.of.flight'].astype(str).str.strip()
purpose_summary['Fatal_Rate_Pct']    = (purpose_summary['Fatal_Rate']*100).round(1)
purpose_summary['Mean_Severity_r']   = purpose_summary['Mean_Severity'].round(2)
purpose_summary['Mean_Damage_r']     = purpose_summary['Mean_Damage'].round(2)
```

```python
phase_summary.to_csv("factor_Broad_phase_of_flight.csv", index=False)
purpose_summary.to_csv("factor_Purpose_of_flight.csv", index=False)
```

# Findings
Grouped results show that severity and fatality are highest in Maneuvering, Climb, and Cruise phases, and lowest during Landing.
By purpose, Instructional and Aerial Application flights are least severe, while Personal and Business flights show higher severity and fatal rates. Accidents during high-energy or unstructured operations are far more severe.
Structured, supervised missions reduce the risk of catastrophic outcomes.

# Recommendations 3
Implement strict Standard Operating Procedures (SOPs), limit early operations to structured mission types (charter, training), and emphasize recurrent training.
This reduces operational risk, increases reliability, and strengthens regulatory compliance for a new aviation division.
